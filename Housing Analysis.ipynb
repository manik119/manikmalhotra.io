```r
#Preparing the model for multiple linear regression - 

model = lm(price~area+bedrooms+bathrooms+stories+mainroad+guestroom+basement+ airconditioning+parking, data = data)
summary(model)
```

The p-values for the coefficients corresponding to Area, Bathroom, Stories, 'Mainroad','Guestroom','Basement','Airconditioning', Parking were significantly lower than the 5% significance levels. We find that an unexpectedly 'Bedroom' doesnt hold sufficient sunificance as the p-value is less than 0.05. Additionally, the R-squared is 64%. Which means that 64% the price change in houses was explained by these variables. The outliers seen in the data contribute to the relative lower accuracy of the model. If we have to increase the accuracy of the model we have identify and remove the outliers in the model.
All the variables have adequate significance. If we have to rank, the variables in terms of significance we it will the following -

```r
#Model evaluation using ANOVA and testing the linear regression model - 
anova(model)

predict(model)
resid(model)

plot(predict(model),resid(model))
```

```r
hist(resid(model),main=paste("HistogramofResiduals"),xlab="Residuals")

par(mfrow=c(2,2))
plot(model)
```

The histogram is not in the perfect bell shape and has bigger tails on both ends. The bigger tails are the result of the outliers present in the data. 
In the 1st scatter plot we observe that residual vs fitted value shown in the red line is not horizontal which suggests that the model's residuals are not well-behaved. Moreover, the distribution of the residual towards the axis also shows that linearity is violated.

```r
#Regression Tree - 
install.packages("tree")
library(tree)

set.seed(5)
train <- sample(1:nrow(data), 0.7 * nrow(data))
trainset <- data[train, ] 
testset <- data[train, ] 

tree.sub <- tree(price~., data = trainset)
summary(tree.sub)
plot(tree.sub)
text(tree.sub,pretty=0)
cv.sub = cv.tree(tree.sub)
cv.sub
```

We can see that the regression tree model has been divided into 3 seperate clusters. We will prune the tree to have a more clear understanding of the model.

```r
#testing MSE pre pruning
predictions <- predict(tree.sub, testset)
actual_values <- testset$price
mse.pre <- mean((predictions - actual_values)^2)
mse.pre
```

```r
#Pruning the tree
plot(cv.sub$size,cv.sub$dev,type="b")
```

```r
prune.sub= prune.tree(tree.sub,best=3)
plot(prune.sub)
text(prune.sub,pretty=0)
```

We used cv.sub for function determining the number nodes that have lowest error and pruned the tree accordingly. As given in the cv.sub graph the tree with 12 terminal nodes will have the lowest error and will the best efficent in understanding the model.

However, when we use 12 terminal nodes we find that the pruned regression tree remains unchanged. This indicates that the pruned tree is the same as the unpruned tree. Here, the tree is split between 3 major sections with area is the primary independent variable followed by bathrooms. 

Regression tree can be used to identify and divide the population into diffrent sections based on the attributes and also give us the price range of the homes in those sections.
